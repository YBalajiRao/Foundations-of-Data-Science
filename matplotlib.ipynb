```python
# 📊 2. Bar Chart — (To compare categories)
# 🤔 Use Case:
# Compare sales of 4 chocolates.

import matplotlib.pyplot as plt
chocolates=['Dairy Milk', 'KitKat', '5 Star', 'Perk']
sales = [40, 70, 30, 50]
plt.bar(chocolates, sales,color='orange')

#labels
plt.xlabel("Chocolate Brand")
plt.ylabel("Sales")
plt.title("Bar chart - Chocolates sales in july")
plt.show()
```

```python
# ✅ 3. Pie Chart — Used To Show Part-to-Whole (% Share)
# 🤔 Use Case:
# You detected different objects underwater, and now you want to see how much each object contributes overall — in percentage.
```

```python
import matplotlib.pyplot as plt
objects = ['Fish', 'Sea Turtles', 'Plastic Waste', 'Coral Reefs', 'Cans/Bottles']
counts = [120, 45, 70, 35, 55]
plt.pie(counts,labels=objects,autopct='%1.1f%%', startangle=140)
plt.title("Percentage of Underwater Objects Detected")
plt.show()
```

```python
# Real-Life Example:
# Suppose 20 students got these marks out of 100:
```

```python
marks = [22, 45, 67, 89, 34, 55, 56, 78, 88, 90, 66, 44, 52, 68, 72, 61, 39, 48, 92, 85]
```

```python
# We want to see how many students are in:

# 0–30

# 30–50

# 50–70

# 70–90

# 90–100

# A histogram is perfect for this.
```

```python
import matplotlib.pyplot as plt
marks = [22, 45, 67, 89, 34, 55, 56, 78, 88, 90, 66, 44, 52, 68, 72, 61, 39, 48, 92, 85]
plt.hist(marks,bins=[10,30,50,70,90,100], color='skyblue',edgecolor='black')
#labels
plt.xlabel('Marks Range')
plt.ylabel('Number of Students')
plt.title('Student Marks Distribution')
plt.show()

```

```python
# What is plt.hist(...) doing?
# This is drawing a Histogram, which is:

# A chart that shows how many values fall inside certain ranges (called bins).
```

```python
# Let's break this into parts:
# marks = [22, 45, 67, 89, 34, 55, 56, 78, 88, 90, 66, 44, 52, 68, 72, 61, 39, 48, 92, 85]
# 👆 These are student marks. Total = 20 students.
# ✅ bins=[0, 30, 50, 70, 90, 100]
# 👉 This splits marks into value ranges like:
# | Bin (Range) | Meaning                        |
# | ----------- | ------------------------------ |
# | `0 to 30`   | How many got between 0 and 30? |
# | `30 to 50`  | How many got 30–50?            |
# | `50 to 70`  | How many got 50–70?            |
# | `70 to 90`  | How many got 70–90?            |
# | `90 to 100` | How many got 90–100?           |

# ➡️ These are the X-axis bars of the histogram.
# ✅ What Python does:
# It goes through each value in marks and counts how many fall in each bin:

# | Bin    | Count                      |
# | ------ | -------------------------- |
# | 0–30   | 1 (22)                     |
# | 30–50  | 5 (45, 34, 44, 39, 48)     |
# | 50–70  | 6 (55, 56, 66, 52, 68, 61) |
# | 70–90  | 6 (78, 88, 72, 85, 89, 90) |
# | 90–100 | 2 (92, 90 again)           |

# (Note: 90 might be double counted based on exact bin rule)
```

```python
import matplotlib.pyplot as plt
detection_times = [3.2, 5.1, 2.7, 6.0, 4.4, 3.5, 5.8, 2.9, 3.1, 4.0, 6.5, 4.7, 5.3, 2.5, 3.8, 5.9, 3.3, 4.2, 6.1, 3.6, 5.0, 4.5, 2.6, 5.7, 6.3]
plt.hist(detection_times,bins=[2.0, 3.5, 5.0, 6.5, 8.0], color='orange', edgecolor='black')
#labels
plt.xlabel('Detection Time (in seconds)')
plt.ylabel('Number of videos')
plt.title("Distribution of underwater Detection Times")
plt.show()
```

```python
# What is a Scatter Plot?
# A scatter plot is used to show the relationship between two numeric variables.

# Each dot (point) represents one data point with:

# X value

# Y value
# 🧠 Simple Real-Life Example:
# Suppose you’re testing your underwater detection system.
# | Video Size (in MB) | Detection Time (in sec) |
# | ------------------ | ----------------------- |
# | 50                 | 2.8                     |
# | 70                 | 3.5                     |
# | 90                 | 4.1                     |
# | 60                 | 3.0                     |
# | 100                | 5.2                     |
# You want to see:

# “Does bigger video size mean more detection time?”

# This is where you use a scatter plot.
```

```python
#syntax 
#plt.scatter(x_values,y_values)
```

```python
# Optional Styling:
# color='red'

# marker='o' (or 'x', '^', '*', etc.)

# s=100 → size of dots
```

```python
import matplotlib.pyplot as plt
video_size=[50,70,90,60,100]
detection_time=[2.8,3.5,4.1,3.0,5.2]
plt.scatter(video_size, detection_time, color='blue', s=100)
#labels
plt.xlabel('Video size (MB)')
plt.ylabel('Detection Time (Sec)')
plt.title('Relation Between video size and Detection Time')
plt.show()
```

```python
#here in above example what we observe is
#🔍 What Each Dot Means:
# | Dot X | Dot Y | Meaning                             |
# | ----- | ----- | ----------------------------------- |
# | 50    | 2.8   | A 50MB video took 2.8 sec to detect |
# | 70    | 3.5   | A 70MB video took 3.5 sec           |
# | 90    | 4.1   | A 90MB video took 4.1 sec           |

# 📌 The X value = First list (video size)
# 📌 The Y value = Second list (detection time)
```
