```python
# Feature	 loc	      iloc
# Full Form	label-based location	integer location
# Works with	Row labels / names	Row numbers / positions
# Use case	Use when you know row labels	Use when you know row numbers
# Supports conditions	‚úÖ Yes	‚ùå No (Only numeric indexing)
# Syntax	df.loc[row_label, column_name]	df.iloc[row_number, column_index]
```
