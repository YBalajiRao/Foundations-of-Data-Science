```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\91944\Desktop\Foundations of Data Science\Linear Regression\housing.csv")
print(df)
```

```python
df.columns
```

```python
# X will have all the features except Price and Address
X = df.drop(['Price', 'Address'], axis=1)

# Y will be the target variable â€” house price
Y = df['Price']

# Print the shape (rows, columns) of X and Y to confirm
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)
```

```python
#first lets import the linear regresio model
from sklearn.linear_model import LinearRegression
```

```python
#Now lets create the Model Object
model=LinearRegression()
```

```python
#Now lets fit the model to Training Data
model.fit(X,Y)
```

```python
#now lets check the slope - coefficients and intercept or bias
print("coeffcients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
predictions=model.predict(X)
print(predictions[:5])
```

```python
custom_input=[[50000,10,4,3,20000]]
model.predict(custom_input)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.hist(Y, bins=30, color='skyblue',edgecolor='black')
plt.title("distribution of house predictions")
plt.xlabel("price")
plt.ylabel("count")
plt.grid(True)
plt.show()
```

```python
predicted_prices = model.predict(X)

plt.figure(figsize=(10, 6))
plt.scatter(Y, predicted_prices, color='green', alpha=0.5)
plt.plot([Y.min(), Y.max()], [Y.min(), Y.max()], color='red', linewidth=2)  # Perfect prediction line
plt.title("Actual vs Predicted House Prices")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Vary only "Avg. Area Income" (1st feature)
incomes = np.linspace(20000, 120000, 100)  # 100 values from 20K to 120K

# Step 2: Keep other features constant: [income, house_age, rooms, bedrooms, population]
other_features = [5, 6, 3, 30000]  # constant

# Step 3: Prepare input data
input_data = [[income, *other_features] for income in incomes]

# Step 4: Predict prices
predicted_prices = model.predict(input_data)

# Step 5: Plot
plt.figure(figsize=(10, 6))
plt.plot(incomes, predicted_prices, color='purple', linewidth=2)
plt.title("Effect of Avg. Area Income on House Price")
plt.xlabel("Avg. Area Income")
plt.ylabel("Predicted House Price")
plt.grid(True)
plt.show()
```
